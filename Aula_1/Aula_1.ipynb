# 🧠 Mapa Mental — Estatística Descritiva

---

## 🎯 1. Tipos de Variáveis

### 📊 Qualitativas
- **Nominal** → sem ordem  
  - Ex: sexo, categoria  
- **Ordinal** → com ordem  
  - Ex: escolaridade, satisfação  

---

### 🔢 Quantitativas
- **Discreta** → contagem  
  - Ex: filhos, gols  
- **Contínua** → medida  
  - Ex: renda, tempo, combustível  

---

## 📍 2. Medidas de Tendência Central

- **Média** → sensível a outliers  
- **Mediana** → robusta a outliers  
- **Moda** → valor mais frequente  

### ⚠️ Regra prática:
- Com outliers → **mediana**
- Sem outliers → **média**

---

## 📏 3. Medidas de Dispersão

- **Desvio padrão** → dispersão em torno da média  
- **Variância** → desvio padrão ao quadrado  
- **Amplitude** → máximo - mínimo  
- **IQR (Intervalo interquartil)** → Q3 - Q1  
- **Coeficiente de Variação (CV)** → variabilidade relativa (%)

---

### 🔥 Coeficiente de Variação (CV)

- Usado para comparar variabilidade entre escalas diferentes  
- Fórmula: CV = desvio padrão / média  

#### Regra prática:
- CV < 20% → baixa variabilidade  
- CV > 30% → alta variabilidade  

---

## 📊 4. Forma da Distribuição

### Assimetria
- Positiva → cauda à direita  
- Negativa → cauda à esquerda  

### Curtose
- Alta → distribuição mais "pontuda"  
- Baixa → distribuição mais "achatada"  

---

## 📦 5. Boxplot

Mostra:
- Mediana  
- Quartis (Q1, Q3)  
- IQR  
- Outliers  

❌ Não mostra a média (por padrão)

### Interpretação:
- Caixa menor → menor variabilidade  
- Caixa maior → maior variabilidade  

---

## 📉 6. Gráficos

- **Histograma** → distribuição de variáveis contínuas  
- **Scatterplot** → relação entre variáveis + outliers  
- **Gráfico de linhas** → tendência ao longo do tempo  
- **Boxplot** → dispersão + outliers  

---

## 🔍 7. Outliers

Afetam:
- Média  
- Desvio padrão  
- Coeficiente de variação  

Não afetam:
- Mediana  

---

## 📈 8. Distribuição Normal

- Simétrica  
- Média = Mediana = Moda  
- Formato de sino  

---

## 📊 9. Intervalo de Confiança

- Parâmetro → fixo  
- Intervalo → variável  

### Interpretação correta:
> Cerca de 95% dos intervalos construídos conterão o valor real do parâmetro

---

## ⚖️ 10. Comparações

### Quando usar CV:
- Comparar grupos com médias diferentes  
- Comparar escalas diferentes  

### Evitar:
- Comparar apenas desvio padrão nesses casos

---

## 🧠 Resumo Final (Prova em 5s)

- Centro → média, mediana, moda  
- Dispersão → desvio, variância, IQR  
- Forma → assimetria, curtose  
- Outlier → afeta média, não mediana  
- Comparação → usar CV  

```python
# 1. IMPORTS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. DADOS EXEMPLO
salarios:list[int] = [22, 25, 28, 30, 32, 35, 38, 40, 120]

# 3. MÉDIA, MEDIANA, MODA
np.mean(salarios), np.median(salarios), pd.Series(salarios).mode() # type: ignore

# 4. DISPERSÃO
np.std(salarios), np.var(salarios)

# 4. DISPERSÃO
np.std(salarios), np.var(salarios)

# 5. COEFICIENTE DE VARIAÇÃO
cv = np.std(salarios) / np.mean(salarios)
cv
```

```python
# 6. BOXPLOT
sns.boxplot(x=salarios)
plt.show() 

# 7. HISTOGRAMA
plt.hist(salarios, bins=10)
plt.show()

# 8. REMOVENDO OUTLIER
salarios_sem_outlier = salarios[:-1]

np.mean(salarios_sem_outlier), np.std(salarios_sem_outlier)
```

```python
# 9. INTERVALO INTERQUARTIL
q1 = np.percentile(salarios, 25)
q3 = np.percentile(salarios, 75)
iqr = q3 - q1

q1, q3, iqr
```

# Aula 1 — SOR vs SOT

## 🧾 SOR (System of Record)
**Definição:** Sistema responsável por registrar a origem e as alterações dos dados.

- **Foco:** Armazenamento, operação e integridade transacional  
- **Função:** Atua como o “contador” do dado (criação e atualizações)  
- **Exemplos:** CRM, ERP  
- **Características:**
  - Pode existir mais de um SOR
  - Normalmente separado por domínio (clientes, financeiro, etc.)

---

## 🎯 SOT (Source of Truth)
**Definição:** Fonte confiável e consolidada dos dados.

- **Foco:** Acurácia, consistência e visão unificada  
- **Função:** Garantir uma única versão da verdade  
- **Características:**
  - Idealmente único por domínio (SSOT — Single Source of Truth)
  - Consolida dados de múltiplos SORs

---

## 🔄 Relação entre SOR e SOT

- **SOR → Origem dos dados**
- **SOT → Consolidação e padronização**

---

## 🏗️ Contextos de Uso

### Engenharia de Dados / Data Lake
- **SOR:** Camada de dados brutos (raw)
- **SOT:** Camada tratada/curada (trusted / refined)

### Negócios
- **SOR:** Sistemas operacionais (ex: CRM de clientes)
- **SOT:** Sistemas de consolidação (ex: MDM)
  - Integra CRM, financeiro, etc.
