#### Como Podemos calcular a probabilidade de ser atendido até um certo tempo?

```python
# Probabilidade de ser atendido em até 8 min na agencia A


# 1. Aplicar a condição: elementos menores ou iguais a 8
elementos_em_ate_08 = sum(1 for x in A if x <= 8)

# 2. Calcular a probabilidade: (casos favoráveis / total de casos)
prob_A = elementos_em_ate_08 / len(A)

print(f"Vetor: {A}")
print(f"Elementos menores ou iguais a 08: {elementos_em_ate_08}")
print(f"Probabilidade de ser atendido na Agência A: {prob_A:.2f} ({prob_A*100}%)")
```

```python
# Probabilidade de ser atendido em até 8 min na agencia B

# Calcular a probabilidade: (casos favoráveis / total de casos)
prob_B = sum(1 for x in B if x < 8) / len(B)

print(f"Vetor: {B}")
print(f"Elementos menores ou iguais a 08: {sum(1 for x in B if x < 8)}")
print(f"Probabilidade de ser atendido na Agencia B: {prob_B:.2f} ({prob_B*100}%)")
```

```python
# Probabilidade de ser atendido em até 8 min na agencia C

# Calcular a probabilidade: (casos favoráveis / total de casos)
prob_C = sum(1 for x in C if x <= 8) / len(B)

print(f"Vetor: {C}")
print(f"Elementos menores ou iguais a 08: {sum(1 for x in C if x <= 8)}")
print(f"Probabilidade de ser atendido na Agencia C: {prob_C:.2f} ({prob_C*100}%)")
```

### Método para resolver problema relacionado a contagem:


 image.png

Exemplificar, encontre todas as combinações possíveis do conjunto {A, B, C, D} com dois elementos:

Listando as combinações com dois elementos, são elas: {A,B}, {A,C}, {A,D}, {B,C}, {B,D} e {C,D}.

Nesse caso é possível ver que há 6 combinações possíveis (só lembrando que os subconjuntos {A,B} e {B,A} são iguais, pois, na combinação, a ordem não é importante)

C(4,2) = 4! /(4-2)!*2!

#### Fatorial
0! = 1
1! = 1
2! = 2 x 1 = 2
3! = 3 x 2 x 1 = 6

```python
import math as math
# Solução

print('3!=',math.factorial(3))
```

```python
n = 60
p= 6
difnp = n - p
print('n!=',math.factorial(n))
print('p!=',math.factorial(p))
print('difnp!=',math.factorial(difnp))
print('combinacoes=',math.factorial(n)/(math.factorial(difnp)*math.factorial(p)))
```

```python
# Cálculo da probabilidade de ganhar com um único jogo da Mega Sena

prob = 1/(math.factorial(n)/(math.factorial(difnp)*math.factorial(p)))
print(prob)
```

#### Sempre há a possibilidade de criar funções

```python
# Permutação: possibilidades de colocação de n objetos em n posições = n!
def permutacao (n):
    return math.factorial(n)

# Arranjo: p objetos em n posições, ordem importa = n!/(n-p)!
def arranjo (n,p):
    return math.factorial(n)/math.factorial(n-p)

# Combinação: p objetos em n posições, ordem não importa = n!/(n-p)!p!
def combinacao (n,p):
    return math.factorial(n)/(math.factorial(n-p)*math.factorial(p))

# Variações possíveis havendo n slots e p possibilidades para cada um
def possibilidades(n,p):
    return p**n
```

```python
print('combinacoes=',combinacao (60,6))
```

```python
# Cálculo da probabilidade de ganhar com um único jogo da Mega Sena

prob = 1/combinacao(60,6)
print(prob)
```

####    Distribuições de probabilidade

    é a função que associa a cada valor possível de uma variável aleatória uma probabilidade.

    Caso a variável seja discreta, teremos uma PMF, caso a variável seja contínua, teremos uma PDF.
    Probability Mass Function (PMF: Função de massa de probabilidade) e
    Probability Density Function (PDF:função de densidade de probabilidade)


####    ATENÇÃO:
    a PMF dá a probabilidade de um ponto do espaço amostral,
    a PDF dá a probabilidade num intervalo, considerando que, para variáveis contínuas, a probabilidade de cada ponto é zero.
    Entretanto, as funções em python que implementam a PDF estimam a probabilidade do ponto pela probabilidade do intervalo infinitesimal.
    
    Função de probabilidade acumulada: Cumulative Distribution Function (CDF)
    É a probabilidade da variável aleatória assumir um valor menor ou igual à x.

    Outras funções de probabilidade
        -Survival function: 1-CDF  ou P(T > t)

        -Interval: pontos da PMF que delimitam um percentual das probabilidades
